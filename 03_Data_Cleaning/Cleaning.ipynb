```python
import numpy as np
import pandas as pd
```

```python
data = {
    'id': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'name': ['Alice', 'bob', 'CHARLIE', 'david', None, 'bob', 'EVE', 'frank', 'Grace', 'henry'],
    'age': [25, 30, None, 22, 28, 30, 'twenty', 35, 27, None],
    'salary': [50000, 60000, 55000, None, 70000, 60000, 80000, None, 65000, 72000],
    'join_date': ['2021-01-15', '2020-03-22', '19-07-2019', None, '2022-11-01', '2020-03-22', '2021/05/10', '2020-08-30', None, '2023-02-14'],
    'department': ['HR', 'IT', 'Finance', 'hr', 'IT', 'IT', None, 'finance', 'HR', 'it']
}
data = pd.DataFrame(data)
print(data)
print(data.shape)
print(data.isnull().sum())
```

```python
#We have to convert twenty into NaN so that we can get mean easily
data['age'] = pd.to_numeric(data['age'],errors='coerce')

#Fill valuess with mean
data['age'] = data['age'].fillna(data['age'].median())
data['salary'] = data['salary'].fillna(data['salary'].median())
data[['age','salary']]
```

```python
# Duplicates remove karo
print("Before:", data.shape)
data = data.drop_duplicates()
print("After:", data.shape)
print(data)
```

```python
data
```

```python
print("Before:", data.shape)
data = data.drop_duplicates(subset='id', keep='first')
print("After:", data.shape)
data
```

```python
# Department lowercase karo
data['department'] = data['department'].str.lower()

# Name title case karo (pehla letter capital)
data['name'] = data['name'].str.title()

print(data[['name', 'department']])
```

```python
data
```

```python
data = data.dropna(subset=['name'])

data['join_date'] = data['join_date'].fillna('unknown')

data['department'] = data['department'].fillna('unknown')

data
```

```python
# Sab dates ek format mein convert karo
data['join_date'] = pd.to_datetime(data['join_date'], errors='coerce')

print(data[['name', 'join_date']])
data
```

```python
print("---CLEANING LOG---")
print(f"1. 'twenty' in age → NaN kiya, mean se fill kiya")
print(f"2. Salary missing values → mean se fill kiya")
print(f"3. Duplicate row (id=5) → remove kiya")
print(f"4. Name → Title Case kiya")
print(f"5. Department → lowercase kiya")
print(f"6. Name None wali row → drop kiya")
print(f"7. Department None → 'unknown' fill kiya")
print(f"8. Join date → datetime convert kiya")
print(f"\n=== FINAL DATASET ===")
print(data)

# CSV save karo
data.to_csv("cleaned_data.csv", index=False)
print("\nCleaned dataset saved!")
```
