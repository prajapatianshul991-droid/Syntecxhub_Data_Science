```python
import pandas as pd
```

```python
# Create a sample dataset
data = {
    'Name':    ['Anshul', 'Ravi', 'Priya', 'Sneha', 'Arjun', 'Neha', 'Rahul', 'Pooja'],
    'Age':     [21, 22, 20, 23, 21, 22, 20, 21],
    'City':    ['Bareilly', 'Delhi', 'Mumbai', 'Pune', 'Lucknow', 'Delhi', 'Bareilly', 'Mumbai'],
    'Maths':   [85, 92, 78, 95, 70, 88, 76, 91],
    'Science': [90, 88, 82, 97, 75, 84, 79, 93],
    'English': [78, 76, 88, 80, 83, 91, 72, 85],
    'Grade':   ['A', 'A', 'B', 'A+', 'B', 'A', 'B', 'A']
}

#Save as Csv
df = pd.DataFrame(data)
df.to_csv('student.csv',index = False)
print("Dataframe is Saved")
df
```
