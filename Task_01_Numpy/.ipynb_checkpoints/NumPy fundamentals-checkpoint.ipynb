```python
import numpy as np
import time
```

```python
#Array Creation
arr1 = np.array([10,2,34,56])
print("1D array:",arr1)
```

```python
arr2 = np.array([[1,2,3],[4,5,6]])
print("2D Array:\n", arr2)
```

```python
#Special Arrays
print("Zeros:\n", np.zeros((3, 3)))
```

```python
print("Ones:\n", np.ones((3, 4)))
```

```python
print("Identity Matrix:\n", np.eye(3))
```

```python
print("arange:", np.arange(0, 20, 2))
```

```python
print("linspace:", np.linspace(0, 1, 5))
```

```python
# Random array (fixed seed for reproducibility)
np.random.seed(42)
rand_arr = np.random.randint(1, 100, size=(4, 4))
print("Random 4x4 Array:\n", rand_arr)
```
