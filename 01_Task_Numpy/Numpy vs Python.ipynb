```python
import numpy as np
import time
```

```python
size = 1_000_000

# Python list
py_list = list(range(size))
start = time.time()
result_list = [x ** 2 for x in py_list]
list_time = time.time() - start

# NumPy array
np_arr = np.arange(size)
start = time.time()
result_np = np_arr ** 2
np_time = time.time() - start

print(f"Python List Time : {list_time:.4f} seconds")
print(f"NumPy Array Time : {np_time:.4f} seconds")
print(f"NumPy is {list_time / np_time:.1f}x FASTER than Python List!")
```
