```python
import numpy as np
```

```python
#Mathematical Operations
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print("a =", a)
print("b =", b)
print("\na + b =", a + b)
print("a * b =", a * b)
print("a**2 =", a**2)
print("b**2 =", b**2)
print("sqrt =", np.sqrt(a))
print("sqrt =", np.sqrt(b))
```

```python
# Matrix multiplication
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])
print("Matrix Multiplication (mat1 @ mat2):\n", mat1 @ mat2)
```

```python
#Axis-wise, and statistical
scores = np.array([[85, 90, 78],
                   [92, 88, 76],
                   [70, 95, 83],
                   [88, 72, 91]])

print("Scores is:\n", scores)
print("\n---Overall Statiscal---")
print("Mean of score:\n", np.mean(scores))
print("Median of score:\n", np.median(scores))
print("Std Dev:\n", np.std(scores).round(2))
print("Min:", np.min(scores), "| Max:",np.max(scores))

print("\n---Axis Wise---")
print("Subject Wise Mean - Axis = 0:", np.mean(scores,axis=0))
print("Student wise Mean - Axis = 1 :", np.mean(scores , axis= 1).round(2))
print("Subject Wise Max - Axis = 0:", np.max(scores,axis=0))
print("Student-wise Sum (axis=1):", np.sum(scores, axis=1))
```
